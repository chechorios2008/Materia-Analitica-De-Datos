<h5> 7. ¿Cuál es el precio promedio? </h5>
<p> El método mean() de Pandas calcula el promedio de una columna de manera predeterminada, ignorando los valores NaN </p> 

```python
promedio_Price = review['price'].mean()
mean_price = promedio_Price.round(2)
print('El precio promedio de los precios es: ', mean_price)
```

<h5> 8. ¿Cuál es el precio más alto pagado? </h5>
<p> El método max() de Pandas recorre la columna "price" y devuelve el valor más alto. La variable precio_maximo almacenará el precio más alto encontrado en la columna. El método max() ignora los valores NaN por defecto. Si hay valores NaN en la columna "price", no se considerarán para calcular el precio máximo.</p>

```python
precio_maximo = review['price'].max()
print('El precio más alto es: ', precio_maximo)
```

<h5> 9. Crea un daframe con todos los vinos de california: </h5>

```python
vinos_california = review.loc[review['province'] == 'California']
vinos_california
```

<h5> 10. Usa idxmax() para encontrar el índice del vino con el precio más alto y luego usa loc para obtener toda la información de ese vino específico. </h5>
<p> Este código primero encuentra el índice del vino con el precio más alto en el DataFrame "vinos_california" utilizando el método idxmax() aplicado a la columna "price". Luego, utiliza la función loc[] para obtener toda la información del vino en ese índice específico.</p>
<p> El método idxmax() en Pandas se utiliza para encontrar el índice de la fila con el valor máximo en una columna o eje específico de un DataFrame.</p>
<p> DataFrame.idxmax ( eje = 0 , skipna = True , numeric_only = False ). Índice de retorno de la primera ocurrencia del máximo sobre el eje solicitado. Se excluyen los valores NA/nulos. los Parámetros descritos en la documentación son opcionales </p>

```python
print("El vino más caro de california es:")
indice_vino_maximo_precio = vinos_california['price'].idxmax()

# Obtener toda la información del vino con el precio más alto
vino_mas_caro = vinos_california.loc[indice_vino_maximo_precio]

print(vino_mas_caro)
```

<h5> 11. Cuáles son los tipos de uva más comunes en California? </h5>

```python
# Crear un DataFrame con variedad de vino y su frecuencia
dataframe_frecuencia_variedad = vinos_california['variety'].value_counts().reset_index()

# Renombrar columnas
dataframe_frecuencia_variedad.columns = ['variedad_vino', 'frecuencia']


dataframe_frecuencia_variedad
```

<p> Identificar las variedades de vinos que tienen una frecuencia mayor a 100, es decir que dentro de los vinos de California están en más de 100 registros.  </p>

```python
vinos_frecuencia_mayor_100 = dataframe_frecuencia_variedad[dataframe_frecuencia_variedad['frecuencia'] > 100]
variedades_con_alta_frecuencia = vinos_frecuencia_mayor_100['variedad_vino'].tolist()
print(variedades_con_alta_frecuencia)
```
