Averigua la suma, la media y la desviación de los puntos marcados por los Boston Celtics (BOS) en toda su historia

```python
# DataFrame con los juegos del equipo Celtics
bos_games = nba[nba['team_id'] == 'BOS']

# Suma, media y desviación de puntos hechos por los Celtics
total_points = bos_games['pts'].sum()
mean_points = bos_games['pts'].mean()
std_point = bos_games['pts'].std()

print("Las estadisticas de los puntos marcados por los Boston Celtics son:")
print(f"El total de puntos marcados es: {total_points}")
print(f"La media de los puntos marcados es: {mean_points}")
print(f"La desviación de los puntos marcados es de: {std_point}")
```

Calcula es total de años que llevan jungando los Celtics. Y luego calcula el promedio de puntos por año


```python
# Crear una nueva columna llamada 'year_game' que indica el año del juego.
bos_games['year_game'] = bos_games['date_game'].str[-4:]
bos_games['year_game'] = bos_games['year_game'].astype(int)
```

```python
# Obtener el año más reciente y el año más antiguo
ultimo_año = bos_games['year_game'].max()
primer_año = bos_games['year_game'].min()

años_jugando = ultimo_año - primer_año + 1

print(f"Años que lleva jugando los Celtics: {años_jugando}")
```

```python
# Agrupar por 'year_game' y calcular el promedio de 'pts'
promedio_pts_por_año = bos_games.groupby('year_game')['pts'].mean()

print("Promedio de puntos por año:")
print(promedio_pts_por_año.round(2))
```


Ahora busca el total de años jugados por los Celtics. Usa la fución `apply` y una función lambda

```python
total_years_played = bos_games['year_game'].apply(lambda x: x).nunique()

print("Total de años jugados:", total_years_played)
```

Ahora que calculaste el numero total de años jugados, calcula el total de puntos por año

```python
total_points_per_year = bos_games.groupby('year_game')['pts'].apply(lambda x: x.sum())
print("Total de puntos por año:\n", total_points_per_year)
```

```python
average_points_per_year = bos_games.groupby('year_game')['pts'].apply(lambda x: x.mean())
print("Promedio de puntos por año:\n", average_points_per_year)
```

Muestra los puntos anotados por los knicks en cada año

```python
knicks_data = nba[nba['fran_id'] == 'Knicks']

# Agrupar por año y sumar los puntos
points_by_year = knicks_data.groupby('year_game')['pts'].sum().reset_index()
points_by_year
```

Grafica este dataframe con la API de pandas usando plot()

```python
# Graficar el dataframe resultante
points_by_year.plot(kind='bar', x='year_game', y='pts', legend=False,figsize=(16, 6))
plt.xlabel('Year')
plt.ylabel('Total Points')
plt.title('Total Points by Year for Knicks')
plt.show()
```

Realiza un gráfico de dispersión de los puntos de los knicks vs los puntos de sus oponentes en los años arriba del año 2000

```python
knicks = nba[(nba['fran_id'] == 'Knicks') & (nba['year_game'] > 2000)]
knicks
```

```python
# Crear el gráfico de dispersión
plt.figure(figsize=(10, 6))
plt.scatter(knicks['pts'], knicks['opp_pts'], alpha=0.7, edgecolors='w', s=100)
plt.xlabel('Puntos de Knicks')
plt.ylabel('Puntos del Oponente')
plt.title('Puntos de Knicks vs Puntos del Oponente')
plt.grid(True)
plt.show()
```

Escoje s = 0.5 para el tamaño de los puntos. Qué observas en x = y ?
<p> Con s = 0.5, los puntos serán mucho más pequeños, por lo que la visualización puede ser más clara si hay muchos puntos cercanos a la línea x = y. Determina si hubo algún empate

Encuenta el número de juegos ganados y perdids por los Miami heats en el año 2013

```python
# Filtrar los juegos de Miami Heat en el año 2013
miami_2013 = nba[(nba['team_id'] == 'MIA') & (nba['year_game'] == 2013)]

# Clasificar los juegos como ganados, perdidos o empatados
miami_2013['result'] = miami_2013.apply(lambda row: 'Win' if row['pts'] > row['opp_pts'] else ('Loss' if row['pts'] < row['opp_pts'] else 'Draw'), axis=1)

result_counts = miami_2013['result'].value_counts()

print("Número de juegos ganados:", result_counts.get('Win', 0))
print("Número de juegos perdidos:", result_counts.get('Loss', 0))
print("Número de juegos empatados:", result_counts.get('Draw', 0))
```

Grafica tu resultado con un grafico de torta

```python
# Crear el gráfico de torta
plt.figure(figsize=(6, 6))
plt.pie(result_counts, labels=result_counts.index, autopct='%1.1f%%', colors=['#4CAF50', '#F44336', '#FFC107'])
plt.title('Resultados de partidos de Miami Heat en 2013')
plt.show()
```
