<p>4. Crea una matriz de 4x4, donde cada elemento es igual a i+j (con i y j siendo el 칤ndice de fila y columna, respectivamente) y calcula su inversa. </p>

```python
# Crear una matriz de 4x4 donde cada elemento es igual a i+j
matrix_4x4 = np.fromfunction(lambda i, j: i + j, (4, 4), dtype=int)

# Calcular la inversa de la matriz
matrix_inverse = np.linalg.inv(matrix_4x4)

print("Matriz original:\n", matrix_4x4)
print("Inversa de la matriz:\n", matrix_inverse)
```

<p> 6.Crea un array de tama침o 3x1 y uno de 1x3, y s칰malos utilizando broadcasting para obtener un array de 3x3. </p>

```python
# Crear un array de tama침o 3x1
array_3x1 = np.array([[1], [2], [3]])

# Crear un array de tama침o 1x3
array_1x3 = np.array([[4, 5, 6]])

# Sumar los arrays utilizando broadcasting para obtener un array de 3x3
result_array = array_3x1 + array_1x3

print("Array 3x1:\n", array_3x1)
print("Array 1x3:\n", array_1x3)
print("Resultado de la suma (3x3):\n", result_array)
```

<h2> Gr치ficos de dispersi칩n, densidad y contorno: </h2>

<p> 11.Genera dos arrays de tama침o 100 con n칰meros aleatorios y crea un gr치fico de dispersi칩n.  </p>

```python
# Generar dos arrays de n칰meros aleatorios
x = np.random.rand(100)
y = np.random.rand(100)

# Crear el gr치fico de dispersi칩n
plt.figure(figsize=(5, 4))
plt.scatter(x, y, alpha=0.7, c='blue', edgecolors='none')
plt.title('Gr치fico de Dispersi칩n')
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.grid(True)
plt.show()
```

<p> 12. Genera un gr치fico de dispersi칩n las variables 洧논 y 洧녽 = 洧멇롐뒳롐(洧논)+ ruido Gaussiano. Donde x es un array con n칰mereos entre -2洧랢 洧녽 2洧랢. Grafica tambi칠n los puntos 洧녽 = 洧멇롐뒳롐(洧논) en el mismo plot  </p>

```python
# Generar valores de x
x = np.linspace(-2*np.pi, 2*np.pi, 100)

# Calcular y = sin(x)
y_sin = np.sin(x)

# Agregar ruido gaussiano a y
ruido = np.random.normal(0, 0.1, size=x.shape)  # ruido gaussiano con media 0 y desviaci칩n est치ndar 0.1
y_con_ruido = y_sin + ruido

# Crear el gr치fico de dispersi칩n y graficar los puntos originales
plt.figure(figsize=(8, 5))
plt.scatter(x, y_con_ruido, label='y = sin(x) + ruido gaussiano', color='blue', alpha=0.7)
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.show()
```

<p> 13. Utiliza la funci칩n np.meshgrid para crear una cuadr칤cula y luego aplica la funci칩n z = np.cos(x) + np.sin(y) para generar y mostrar un gr치fico de contorno. </p>

```python
# Definir el rango de valores para x e y
x = np.linspace(-2*np.pi, 2*np.pi, 100)
y = np.linspace(-2*np.pi, 2*np.pi, 100)

# Crear la cuadr칤cula utilizando np.meshgrid
X, Y = np.meshgrid(x, y)

# Calcular z = cos(x) + sin(y)
Z = np.cos(X) + np.sin(Y)

# Crear el gr치fico de contorno
plt.figure(figsize=(7, 5))
contour = plt.contour(X, Y, Z, cmap='viridis')
plt.colorbar(contour, label='z = cos(x) + sin(y)')
plt.title('Gr치fico de Contorno: z = cos(x) + sin(y)')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.show()
```

<p> 19. Experimenta con diferentes valores de bins (por ejemplo, 10, 30, 50) en un histograma y observa c칩mo cambia la representaci칩n. </p>

```python
# Generar datos con distribuci칩n normal
datos = np.random.normal(loc=0, scale=1, size=1000)

# Crear el histograma con diferentes valores de bins
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.hist(datos, bins=10, alpha=0.7, edgecolor='black')
plt.title('Histograma (bins=10)')
plt.xlabel('Valores')
plt.ylabel('Frecuencia')
plt.grid(True)

plt.subplot(1, 3, 2)
plt.hist(datos, bins=30, alpha=0.7, edgecolor='black')
plt.title('Histograma (bins=30)')
plt.xlabel('Valores')
plt.ylabel('Frecuencia')
plt.grid(True)

plt.subplot(1, 3, 3)
plt.hist(datos, bins=50, alpha=0.7, edgecolor='black')
plt.title('Histograma (bins=50)')
plt.xlabel('Valores')
plt.ylabel('Frecuencia')
plt.grid(True)

plt.tight_layout()
plt.show()
```

<p> 20. A침ade una l칤nea vertical que indique la media de los datos en el histograma. </p>

```python
# Generar datos con distribuci칩n normal
datos = np.random.normal(loc=0, scale=1, size=1000)

plt.figure(figsize=(15, 5))

for i, bins in enumerate([10, 30, 50], start=1):
    plt.subplot(1, 3, i)
    plt.hist(datos, bins=bins, alpha=0.7, edgecolor='black')
    
    media = np.mean(datos)
    
    # A침adir l칤nea vertical para la media
    plt.axvline(media, color='red', linestyle='dashed', linewidth=1.5, label=f'Media = {media:.2f}')
    
    plt.title(f'Histograma (bins={bins})')
    plt.xlabel('Valores')
    plt.ylabel('Frecuencia')
    plt.legend()
    plt.grid(True)

plt.tight_layout()
plt.show()
```
