<p>4. Crea una matriz de 4x4, donde cada elemento es igual a i+j (con i y j siendo el índice de fila y columna, respectivamente) y calcula su inversa. </p>

```python
# Crear una matriz de 4x4 donde cada elemento es igual a i+j
matrix_4x4 = np.fromfunction(lambda i, j: i + j, (4, 4), dtype=int)

# Calcular la inversa de la matriz
matrix_inverse = np.linalg.inv(matrix_4x4)

print("Matriz original:\n", matrix_4x4)
print("Inversa de la matriz:\n", matrix_inverse)
```

<h2> Broadcasting e indexado de Arrays: </h2>

<p> 6.Crea un array de tamaño 3x1 y uno de 1x3, y súmalos utilizando broadcasting para obtener un array de 3x3. </p>

```python
# Crear un array de tamaño 3x1
array_3x1 = np.array([[1], [2], [3]])

# Crear un array de tamaño 1x3
array_1x3 = np.array([[4, 5, 6]])

# Sumar los arrays utilizando broadcasting para obtener un array de 3x3
result_array = array_3x1 + array_1x3

print("Array 3x1:\n", array_3x1)
print("Array 1x3:\n", array_1x3)
print("Resultado de la suma (3x3):\n", result_array)
```

<h2> Gráficos de dispersión, densidad y contorno: </h2>

<p> 11.Genera dos arrays de tamaño 100 con números aleatorios y crea un gráfico de dispersión.  </p>

```python
# Generar dos arrays de números aleatorios
x = np.random.rand(100)
y = np.random.rand(100)

# Crear el gráfico de dispersión
plt.figure(figsize=(5, 4))
plt.scatter(x, y, alpha=0.7, c='blue', edgecolors='none')
plt.title('Gráfico de Dispersión')
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.grid(True)
plt.show()
```

<p> 12. Genera un gráfico de dispersión las variables 𝑥 y 𝑦 = 𝑠𝑖𝑛(𝑥)+ ruido Gaussiano. Donde x es un array con númereos entre -2𝜋 𝑦 2𝜋. Grafica también los puntos 𝑦 = 𝑠𝑖𝑛(𝑥) en el mismo plot  </p>

```python
# Generar valores de x
x = np.linspace(-2*np.pi, 2*np.pi, 100)

# Calcular y = sin(x)
y_sin = np.sin(x)

# Agregar ruido gaussiano a y
ruido = np.random.normal(0, 0.1, size=x.shape)  # ruido gaussiano con media 0 y desviación estándar 0.1
y_con_ruido = y_sin + ruido

# Crear el gráfico de dispersión y graficar los puntos originales
plt.figure(figsize=(8, 5))
plt.scatter(x, y_con_ruido, label='y = sin(x) + ruido gaussiano', color='blue', alpha=0.7)
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.show()
```

<p> 13. Utiliza la función np.meshgrid para crear una cuadrícula y luego aplica la función z = np.cos(x) + np.sin(y) para generar y mostrar un gráfico de contorno. </p>

```python
# Definir el rango de valores para x e y
x = np.linspace(-2*np.pi, 2*np.pi, 100)
y = np.linspace(-2*np.pi, 2*np.pi, 100)

# Crear la cuadrícula utilizando np.meshgrid
X, Y = np.meshgrid(x, y)

# Calcular z = cos(x) + sin(y)
Z = np.cos(X) + np.sin(Y)

# Crear el gráfico de contorno
plt.figure(figsize=(7, 5))
contour = plt.contour(X, Y, Z, cmap='viridis')
plt.colorbar(contour, label='z = cos(x) + sin(y)')
plt.title('Gráfico de Contorno: z = cos(x) + sin(y)')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.show()
```

<p> 15. A partir de la misma función del ejercicio 12, genera un gráfico de contorno lleno.  </p>

```python
# Definir el rango de valores para x e y
x = np.linspace(-2*np.pi, 2*np.pi, 100)
y = np.linspace(-2*np.pi, 2*np.pi, 100)

# Crear la cuadrícula utilizando np.meshgrid
X, Y = np.meshgrid(x, y)

# Calcular z = cos(x) + sin(y)
Z = np.cos(X) + np.sin(Y)

plt.figure(figsize=(7, 5))
contour = plt.contourf(X, Y, Z, cmap='viridis')
plt.colorbar(contour, label='z = cos(x) + sin(y)')
plt.title('Gráfico de Contorno Lleno: z = cos(x) + sin(y)')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.show()
```

<h2> Histogramas: </h2>

<p> 17. Crea un histograma a partir de un array de 1000 números aleatorios generados con una distribución normal. </p>

```python
# Generar 1000 números aleatorios con distribución normal
datos = np.random.randn(1000)

plt.figure(figsize=(8, 6))
plt.hist(datos, bins=30, edgecolor='black', alpha=0.7)
plt.title('Histograma de Distribución Normal')
plt.xlabel('Valores')
plt.ylabel('Frecuencia')
plt.grid(True)
plt.show()
```

<p> 18. Genera dos sets de datos con distribuciones normales diferentes y muéstralos en el mismo histograma.  </p>

```python
# Generar dos conjuntos de datos con distribuciones normales diferentes
datos1 = np.random.normal(loc=0, scale=1, size=1000)  # media=0, desviación estándar=1
datos2 = np.random.normal(loc=2, scale=1.5, size=1000)  # media=2, desviación estándar=1.5

plt.figure(figsize=(7, 5))
plt.hist(datos1, bins=30, alpha=0.7, label='Datos 1', edgecolor='black')
plt.hist(datos2, bins=30, alpha=0.7, label='Datos 2', edgecolor='black')
plt.title('Histograma de Distribuciones Normales')
plt.xlabel('Valores')
plt.ylabel('Frecuencia')
plt.legend()
plt.grid(True)
plt.show()
```

<p> 19. Experimenta con diferentes valores de bins (por ejemplo, 10, 30, 50) en un histograma y observa cómo cambia la representación. </p>

```python
# Generar datos con distribución normal
datos = np.random.normal(loc=0, scale=1, size=1000)

# Crear el histograma con diferentes valores de bins
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.hist(datos, bins=10, alpha=0.7, edgecolor='black')
plt.title('Histograma (bins=10)')
plt.xlabel('Valores')
plt.ylabel('Frecuencia')
plt.grid(True)

plt.subplot(1, 3, 2)
plt.hist(datos, bins=30, alpha=0.7, edgecolor='black')
plt.title('Histograma (bins=30)')
plt.xlabel('Valores')
plt.ylabel('Frecuencia')
plt.grid(True)

plt.subplot(1, 3, 3)
plt.hist(datos, bins=50, alpha=0.7, edgecolor='black')
plt.title('Histograma (bins=50)')
plt.xlabel('Valores')
plt.ylabel('Frecuencia')
plt.grid(True)

plt.tight_layout()
plt.show()
```

<p> 20. Añade una línea vertical que indique la media de los datos en el histograma. </p>

```python
# Generar datos con distribución normal
datos = np.random.normal(loc=0, scale=1, size=1000)

plt.figure(figsize=(15, 5))

for i, bins in enumerate([10, 30, 50], start=1):
    plt.subplot(1, 3, i)
    plt.hist(datos, bins=bins, alpha=0.7, edgecolor='black')
    
    media = np.mean(datos)
    
    # Añadir línea vertical para la media
    plt.axvline(media, color='red', linestyle='dashed', linewidth=1.5, label=f'Media = {media:.2f}')
    
    plt.title(f'Histograma (bins={bins})')
    plt.xlabel('Valores')
    plt.ylabel('Frecuencia')
    plt.legend()
    plt.grid(True)

plt.tight_layout()
plt.show()
```
